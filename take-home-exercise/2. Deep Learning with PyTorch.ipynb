## Logistic Regression as a 1-Layer Neural Network

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Introduction</h3>
</div>

#### Helper Function: Sigmoid

````python
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 1 - Initializing parameters</h3>
</div>

````python
def initialize_parameters(dim):    
    w = np.zeros((dim,1))
    b = 0
    
    parameters = {"w": w,
                  "b": b}
    return parameters
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Steps 2.1 & 2.2 - Forward and Backward propagation</h3>
</div>

````python
def propagate(parameters, X, Y):
    #Number of examples in training set
    m = X.shape[1]
    
    #Retrieve parameters
    w = parameters["w"]
    b = parameters["b"]
    
    ## Forward Propagation (from X to cost)
    # Compute sigmoid activation
    A = sigmoid(np.dot(w.T,X) + b) 
    ## Negative log-likelihood cost for logistic regression
    cost = np.sum(((- np.log(A))*Y + (-np.log(1-A))*(1-Y)))/m
    
    ## Backward Propagation (to find gradient)
    # Gradient of loss with respect to w
    dw = (np.dot(X,(A-Y).T))/m
    # Gradient of loss with respect to b
    db = (np.sum(A-Y))/m
    
    # Return cost and gradients ; [[21]] --> 21
    cost = np.squeeze(cost)    
    grads = {"dw": dw,
             "db": db}
    
    return grads, cost
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.3 - Update Parameters</h3>
</div>

````python
def update_parameters(parameters, grads, learning_rate):
    
    # Retrieve w, b from parameters
    w = parameters["w"]
    b = parameters["b"]
    
    # Retrieve derivatives from grads
    dw = grads["dw"]
    db = grads["db"]
    
    # Update rule for each parameter
    w = w - (learning_rate*dw)
    b = b - (learning_rate*db)
    
    parameters = {"w": w,
                  "b": b}
    
    return parameters
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 3 - Combine into optimize() function</h3>
</div>

````python
#Building logistic regression model using previously defined functions
def optimize(X, Y, learning_rate = 0.01, num_iterations = 10000, print_cost = False):

    # Initialize parameters
    parameters = initialize_parameters(X.shape[0])
    
    # Loop (gradient descent)
    # List of all costs computed during optimization used to plot the learning curve
    costs = []
    
    for i in range(num_iterations):
        
        # Cost and gradient calculation
        grads, cost = propagate(parameters, X, Y)
    
        # Update parameters
        parameters = update_parameters(parameters, grads, learning_rate)
        
        # Record the costs
        if i % 100 == 0:
            costs.append(cost)
        
        # Print the cost every 100 training iterations
        if print_cost and i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
            
    # Returns parameters learnt by the model. They can then be used to predict output
    return parameters, grads, costs
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 4 - Computing Predictions</h3>
</div>

````python
def predict(parameters, X):
    
    #Retrieve parameters
    w = parameters["w"]
    b = parameters["b"]
    
    m = X.shape[1]
    Y_prediction = np.zeros((1,m))
    w = w.reshape(X.shape[0], 1)
    
    # Compute vector "A" predicting the probabilities of the class being positive, with dimensions (1,m)
    A = sigmoid(np.dot(w.T,X) + b)           
    
    Y_prediction = (A >= 0.5) * 1.0
    
    assert(Y_prediction.shape == (1, m))
    
    return Y_prediction
````

````python
#Building logistic regression model using previously defined functions
def model(X_train, Y_train, X_test, Y_test, num_iterations = 10000, learning_rate = 0.01, print_cost = False):

    # Optimization
    parameters, grads, costs = optimize(X_train, Y_train, learning_rate, num_iterations, print_cost)
    
    # Predict test/train set examples
    Y_prediction_test = predict(parameters, X_test)
    Y_prediction_train = predict(parameters, X_train)

    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "parameters" : parameters, 
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
````

## 2-Layer Neural Network

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Introduction</h3>
</div>

#### Helper Function: Sigmoid

Let's compute $sigmoid( w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$ to make predictions.

````python
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
````

#### Defining the neural network structure ####
- n_x: the size of the input layer
- n_h: the size of the hidden layer (set this to 4) 
- n_y: the size of the output layer

````python
def layer_sizes(X, Y):
    
    #Size of input layer
    n_x = X.shape[0]
    #Size of hidden layer
    n_h = 4
    #Size of output layer
    n_y = Y.shape[0]
    
    return (n_x, n_h, n_y)
````

````python
layer_sizes(train_set_x, train_set_y)
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 1 - Initializing parameters</h3>
</div>

Randomly initialize weights matrices with random values and bias vectors as zeros.

````python
def initialize_parameters(n_x, n_h, n_y):
    #Set up seed to match random initialization
    np.random.seed(1)
    
    W1 = np.random.randn(n_h,n_x) * 0.01
    b1 = np.zeros((n_h,1))
    W2 = np.random.randn(n_y,n_h) * 0.01
    b2 = np.zeros((n_y,1))
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.1 - Forward Propagation</h3>
</div>

````python
def forward_propagation(X, parameters):

    # Retrieve each parameter from the dictionary "parameters"
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # Implement Forward Propagation to calculate A2 (probabilities)
    Z1 = np.dot(W1,X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2,A1) + b2
    A2 = sigmoid(Z2)

    assert(A2.shape == (1, X.shape[1]))
    
    # Values needed in the backpropagation are stored in "cache". This will be given as an input to the backpropagation
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache
````

````python
def compute_cost(A2, Y, parameters):

    #Number of examples
    m = Y.shape[1]

    #Compute the cross-entropy cost
    logprobs = np.multiply(Y ,np.log(A2)) + np.multiply((1-Y), np.log(1-A2))
    cost = (-1/m) * np.sum(logprobs)
    
    cost = float(np.squeeze(cost)) 
                                    
    assert(isinstance(cost, float))
    
    return cost
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.2 - Backward Propagation</h3>
</div>

````python
def backward_propagation(parameters, cache, X, Y):
    
    # Number of examples
    m = X.shape[1]
    
    # First, retrieve W1 and W2 from the dictionary "parameters".
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
        
    # Retrieve also A1 and A2 from dictionary "cache".
    A1 = cache["A1"]
    A2 = cache["A2"]
    Z1 = cache["Z1"]
    Z2 = cache["Z2"]
    
    # Backward propagation: calculate dW1, db1, dW2, db2. 
    dZ2 = A2 - Y
    dW2 = (1/m) * np.dot(dZ2,A1.T)
    db2 = (1/m) *(np.sum(dZ2,axis=1,keepdims=True))
    dZ1 = np.dot(W2.T,dZ2) * (1 - np.power(A1,2))
    dW1 = (1/m) *(np.dot(dZ1,X.T))
    db1 = (1/m) *(np.sum(dZ1, axis=1, keepdims=True))
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    
    return grads
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.3 - Update Parameters</h3>
</div>

````python
def update_parameters(parameters, grads, learning_rate):
    
    # Retrieve each parameter from the dictionary "parameters"
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # Retrieve each gradient from the dictionary "grads"
    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]
    
    # Update rule for each parameter
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 3 - Combine into optimize() function</h3>
</div>

````python
def optimize(X, Y, n_h, learning_rate=0.01, num_iterations = 10000, print_cost=False):
    n_x = layer_sizes(X, Y)[0]
    n_y = layer_sizes(X, Y)[2]
    
    # Initialize parameters
    parameters = initialize_parameters(n_x, n_h, n_y)
    
    # Loop (gradient descent)
    costs = []
    for i in range(0, num_iterations):
        # Forward propagation
        A2, cache = forward_propagation(X, parameters)
        # Cost function
        cost = compute_cost(A2, Y, parameters)
        # Backpropagation
        grads = backward_propagation(parameters, cache, X, Y)
        # Update rule for each parameter
        parameters = update_parameters(parameters, grads, learning_rate)
        
        # Record the costs
        if i % 100 == 0:
            costs.append(cost)
        # If print_cost=True, Print the cost every 1000 iterations
        if print_cost and i % 1000 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
    
    # Returns parameters learnt by the model. They can then be used to predict output
    return parameters, grads, costs
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 4 - Computing Predictions</h3>
</div>

**Reminder**: predictions = $y_{prediction} = \mathbb 1 \text{{activation > 0.5}} = \begin{cases}
      1 & \text{if}\ activation > 0.5 \\
      0 & \text{otherwise}
    \end{cases}$  
    
As an example, if you would like to set the entries of a matrix X to 0 and 1 based on a threshold you would do: ```X_new = (X > threshold)```

````python
def predict(parameters, X):
    #Number of examples
    m = X.shape[1]
    
    # Computes probabilities using forward propagation, and classifies to 0/1 using 0.5 as the threshold.
    A2, cache = forward_propagation(X, parameters)
    Y_prediction = (A2 > 0.5)
    
    assert(Y_prediction.shape == (1, m))

    return Y_prediction
````

````python
#Building a 2-Layer neural network model using previously defined functions
def model(X_train, Y_train, X_test, Y_test, n_h, num_iterations = 10000, learning_rate = 0.01, print_cost = False):

    # Optimization
    parameters, grads, costs = optimize(X_train, Y_train, n_h, learning_rate, num_iterations, print_cost)
    
    # Predict test/train set examples
    Y_prediction_test = predict(parameters, X_test)
    Y_prediction_train = predict(parameters, X_train)

    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    
    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "parameters" : parameters, 
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
````

## L-Layer Neural Network

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Introduction</h3>
</div>

#### Helper Functions

````python
def sigmoid(Z):
    A = 1/(1+np.exp(-Z))
    cache = Z
    
    return A, cache
````

````python
def relu(Z):
    A = np.maximum(0,Z)
    
    assert(A.shape == Z.shape)
    
    cache = Z 
    return A, cache
````

````python
def relu_backward(dA, cache):
    #dA is post-activation gradient and dZ is gradient of cost w.r.t. Z
    Z = cache
    dZ = np.array(dA, copy=True)
    
    dZ[Z <= 0] = 0
    
    assert (dZ.shape == Z.shape)
    
    return dZ
````

````python
def sigmoid_backward(dA, cache):
    #dA is post-activation gradient and dZ is gradient of cost w.r.t. Z
    Z = cache
    
    s = 1/(1+np.exp(-Z))
    dZ = dA * s * (1-s)
    
    assert (dZ.shape == Z.shape)
    
    return dZ
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 1 - Initializing parameters</h3>
</div>

````python
def initialize_parameters(layer_dims):
    """
    Arguments:
    layer_dims -- python array (list) containing the dimensions of each layer in our network
    
    Returns:
    parameters -- python dictionary containing your parameters "W1", "b1", ..., "WL", "bL":
                    Wl -- weight matrix of shape (layer_dims[l], layer_dims[l-1])
                    bl -- bias vector of shape (layer_dims[l], 1)
    """
    
    np.random.seed(1)
    parameters = {}
    
    # Number of layers in the network
    L = len(layer_dims)            

    for l in range(1, L):
        parameters['W' + str(l)] = np.random.randn(layer_dims[l],layer_dims[l-1]) * 0.01
        parameters['b' + str(l)] = np.zeros((layer_dims[l],1))

        assert(parameters['W' + str(l)].shape == (layer_dims[l], layer_dims[l-1]))
        assert(parameters['b' + str(l)].shape == (layer_dims[l], 1))

    return parameters
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.1 - Forward Propagation</h3>
</div>

````python
def forward_propagation(X, parameters):
    """
    Implement forward propagation for the [LINEAR->RELU]*(L-1)->LINEAR->SIGMOID computation
    
    Arguments:
    X -- data, numpy array of shape (input size, number of examples)
    parameters -- output of initialize_parameters_deep()
    
    Returns:
    AL -- last post-activation value
    caches -- list of caches containing:
                every cache of linear_activation_forward() (there are L-1 of them, indexed from 0 to L-1)
    """

    caches = []
    A = X
    # Number of layers in the neural network
    L = len(parameters) // 2                  
    
    for l in range(1, L+1):
        A_prev = A 
        
        W = parameters['W' + str(l)]
        b = parameters['b' + str(l)]
        
        Z = np.dot(W,A_prev) + b

        assert(Z.shape == (parameters['W' + str(l)].shape[0], A_prev.shape[1]))
        linear_cache = (A_prev, W, b)
        
        # Implement [LINEAR -> RELU]*(L-1).
        if l < L:
            A, activation_cache = relu(Z)
            
        # Implement LINEAR -> SIGMOID.

        elif l == L:
            Yhat, activation_cache = sigmoid(Z)
        
        #Add "cache" to the "caches" list.

        cache = (linear_cache, activation_cache)
        caches.append(cache)
    
    assert(Yhat.shape == (1,X.shape[1]))

    return Yhat, caches
````

````python
def compute_cost(Yhat, Y):
    # Number of examples
    m = Y.shape[1]

    # Compute loss
    cost = (-1/m) * (np.dot(Y, np.log(Yhat).T) + np.dot((1-Y), np.log(1-Yhat).T))
    
    #This turns [[21]] into 21
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    
    return cost
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.2 - Backward Propagation</h3>
</div>

````python
def backward_propagation(Yhat, Y, caches):
    """
    Implement the backward propagation for the [LINEAR->RELU] * (L-1) -> LINEAR -> SIGMOID group
    
    Arguments:
    Yhat -- probability vector, output of the forward propagation (L_model_forward())
    Y -- true "label" vector (containing 0 if non-cat, 1 if cat)
    caches -- list of caches containing:
                every cache of linear_activation_forward() with "relu" (it's caches[l], for l in range(L-1) i.e l = 0...L-2)
                the cache of linear_activation_forward() with "sigmoid" (it's caches[L-1])
    
    Returns:
    grads -- A dictionary with the gradients
             grads["dYhat" + str(l)] = ... 
             grads["dW" + str(l)] = ...
             grads["db" + str(l)] = ... 
    """
    grads = {}
    # Number of layers
    L = len(caches)
    m = Yhat.shape[1]
    Y = Y.reshape(Yhat.shape)
    
    # Initializing the backpropagation
    dYhat = - (np.divide(Y, Yhat) - np.divide(1 - Y, 1 - Yhat))
    
    # Loop from l=L-1 to l=0
    for l in reversed(range(L)):
        current_cache = caches[l]
        linear_cache, activation_cache = current_cache

        # Lth layer (SIGMOID -> LINEAR) gradients
        if l == (L-1):                    
            dZ = sigmoid_backward(dYhat, activation_cache)

        # lth layer: (RELU -> LINEAR) gradients.
        elif l < (L-1):
            dZ = relu_backward(grads["dA" + str(l + 1)], activation_cache)
            
        
        A_prev, W, b = linear_cache
        m = A_prev.shape[1]
        
        grads["dA" + str(l)] = np.dot(W.T,dZ)
        grads["dW" + str(l + 1)] = (1/m) * np.dot(dZ, A_prev.T)
        grads["db" + str(l + 1)] = (1/m) * np.sum(dZ, axis=1, keepdims=True)
        
    return grads
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 2.3 - Update Parameters</h3>
</div>

````python
def update_parameters(parameters, grads, learning_rate):
    # Number of layers in the neural network
    L = len(parameters) // 2 

    # Update rule for each parameter. Use a for loop.
    for l in range(L):
        parameters["W" + str(l+1)] = parameters["W" + str(l+1)] - learning_rate * grads["dW" + str(l+1)]
        parameters["b" + str(l+1)] = parameters["b" + str(l+1)] - learning_rate * grads["db" + str(l+1)]
    return parameters
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 3 - Combine into optimize() function</h3>
</div>

````python
def optimize(X, Y, layers_dims, learning_rate=0.01, num_iterations = 10000, print_cost=False):
    
    np.random.seed(1)
    costs = [] 
    
    # Parameters initialization
    parameters = initialize_parameters(layers_dims)
    
    # Loop (gradient descent)
    for i in range(0, num_iterations):

        # Forward propagation: [LINEAR -> RELU]*(L-1) -> LINEAR -> SIGMOID
        Yhat, caches = forward_propagation(X, parameters)
        
        # Compute cost.
        cost = compute_cost(Yhat, Y)
    
        # Backward propagation
        grads = backward_propagation(Yhat, Y, caches)
 
        # Update parameters
        parameters = update_parameters(parameters, grads, learning_rate)
                
        # Print the cost every 100 training example
        if print_cost and i % 1000 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
        if i % 100 == 0:
            costs.append(cost)

    # Returns parameters learnt by the model. They can then be used to predict output
    return parameters, grads, costs
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Step 4 - Computing Predictions</h3>
</div>

````python
def predict(parameters, X):
    
    m = X.shape[1]
    # Number of layers in the neural network
    n = len(parameters) // 2 
    
    # Forward propagation
    A, caches = forward_propagation(X, parameters)
    Y_prediction = (A > 0.5)
    
    assert(Y_prediction.shape == (1, m))

    return Y_prediction
````

````python
#Building an L-Layer neural network model using previously defined functions
def model(X_train, Y_train, X_test, Y_test, layers_dims, num_iterations = 10000, learning_rate = 0.01, print_cost = False):

    # Optimization
    parameters, grads, costs = optimize(X_train, Y_train, layers_dims, learning_rate, num_iterations, print_cost)
    
    # Predict test/train set examples
    Y_prediction_test = predict(parameters, X_test)
    Y_prediction_train = predict(parameters, X_train)

    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    
    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "parameters" : parameters, 
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
````

````python
#Train data
class data_train(Dataset):
    
    def __init__(self, data_X, data_y):
        self.data_X = data_X
        self.data_y = data_y
        
    def __getitem__(self, index):
        return self.data_X[index], self.data_y[index]
        
    def __len__ (self):
        return len(self.data_X)
````

````python
#Test data    
class data_test(Dataset):
    
    def __init__(self, data_X):
        self.data_X = data_X
        
    def __getitem__(self, index):
        return self.data_X[index]
        
    def __len__ (self):
        return len(self.data_X)
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Helper Functions for Classification</h3>
</div>

````python
#Calculate accuracy
def binary_accuracy(y_pred, y_test):
    y_pred_label = torch.round(torch.sigmoid(y_pred))

    sum_correct_results = (y_pred_label == y_test).sum().float()
    accuracy = sum_correct_results/y_test.shape[0]
    accuracy = torch.round(accuracy * 100)
    
    return accuracy
````

````python
#Predict on training data
def get_predictions(model, loader, train=False):
    y_pred_list = []
    model.eval()
    with torch.no_grad():
        if train:
            for X_batch, _ in loader:
                X_batch = X_batch.to(device)
                y_pred = model(X_batch)
                y_pred = torch.sigmoid(y_pred)
                y_pred_label = torch.round(y_pred)
                y_pred_list.append(y_pred_label.cpu().numpy())
        else:
            for X_batch in loader:
                X_batch = X_batch.to(device)
                y_pred = model(X_batch)
                y_pred = torch.sigmoid(y_pred)
                y_pred_label = torch.round(y_pred)
                y_pred_list.append(y_pred_label.cpu().numpy())

    y_pred_list = [x.squeeze().tolist() for x in y_pred_list]
    y_pred_list = [val for inner_list in y_pred_list for val in inner_list]
    return(y_pred_list)
````

## 5-Layer Neural Network for Regression in PyTorch

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Define Architecture and Hyperparameters</h3>

````python
#Define Architecture 
num_features = train_set_x_reg.shape[1]
hidden_layers = [128,256,256,256]
num_classes = train_set_y.reshape(-1,1).shape[1]
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Initialize Dataloader</h3>

````python
train_data = data_train(torch.FloatTensor(train_set_x_reg), 
                       torch.FloatTensor(train_set_y_reg))

test_data = data_test(torch.FloatTensor(test_set_x_reg))
````

````python
train_loader = DataLoader(dataset=train_data, 
                          batch_size=batch_size,
                          shuffle=True)

test_loader = DataLoader(dataset=test_data,
                         batch_size=batch_size,
                         shuffle=False)
````

<div class="alert alert-info" style="background-color:#6C5B7B; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Initialize Model</h3>

````python
# Fully connected neural network with four hidden layers
class FiveLayerNeuralNet(nn.Module):
    def __init__(self, num_features, hidden_layers, num_classes):
        super(FiveLayerNeuralNet, self).__init__()
        self.fc1 = nn.Linear(num_features, hidden_layers[0]) 
        self.fc2 = nn.Linear(hidden_layers[0], hidden_layers[1]) 
        self.fc3 = nn.Linear(hidden_layers[1], hidden_layers[2]) 
        self.fc4 = nn.Linear(hidden_layers[2], hidden_layers[3]) 
        self.fc5 = nn.Linear(hidden_layers[3], num_classes)  
        
        self.relu = nn.ReLU()

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)
        out = self.relu(out)
        out = self.fc4(out)
        out = self.relu(out)
        out = self.fc5(out)
        return out

model = FiveLayerNeuralNet(num_features, hidden_layers, num_classes).to(device)
print(model)
````

## Helper Functions and Configuration

<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Helper Classes</h3>
</div>

````python
class Dictionary(object):
    def __init__(self):
        self.word2idx = {}
        self.idx2word = {}
        self.idx = 0
    
    def add_word(self, word):
        if not word in self.word2idx:
            self.word2idx[word] = self.idx
            self.idx2word[self.idx] = word
            self.idx += 1
    
    def __len__(self):
        return len(self.word2idx)
````

````python
class Corpus(object):
    def __init__(self):
        self.dictionary = Dictionary()

    def get_data(self, path, batch_size=20):
        # Add words to the dictionary
        with open(path, 'r') as f:
            tokens = 0
            for line in f:
                words = line.split() + ['<eos>']
                tokens += len(words)
                for word in words: 
                    self.dictionary.add_word(word)  
        
        # Tokenize the file content
        ids = torch.LongTensor(tokens)
        token = 0
        with open(path, 'r') as f:
            for line in f:
                words = line.split() + ['<eos>']
                for word in words:
                    ids[token] = self.dictionary.word2idx[word]
                    token += 1
        num_batches = ids.size(0) // batch_size
        ids = ids[:num_batches*batch_size]
        return ids.view(batch_size, -1)
````

## Training Model

<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Define Architecture and Hyperparameters</h3>
</div>

````python
#Define Hyperparameters 
embed_size = 128
hidden_size = 1024
num_layers = 1
num_samples = 1000     # number of words to be sampled
seq_length = 30
num_epochs = 5
#Set batch size for defining variant of gradient descent; train_set_x.shape[0] is for batch gradient descent
batch_size = 20
learning_rate = 0.001
````

<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Load Dataset</h3>
</div>

````python
# Load "Penn Treebank" dataset
corpus = Corpus()
ids = corpus.get_data('datasets/langmdldata.txt', batch_size)
vocab_size = len(corpus.dictionary)
num_batches = ids.size(1) // seq_length
````

<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Initialize Model</h3>
</div>

````python
class RNN(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size, num_layers):
        super(RNN, self).__init__()
        self.embed = nn.Embedding(vocab_size, embed_size)
        self.lstm = nn.LSTM(embed_size, hidden_size, num_layers, batch_first=True)
        self.linear = nn.Linear(hidden_size, vocab_size)
        
    def forward(self, x, h):
        # Embed word ids to vectors
        x = self.embed(x)
        
        # Forward propagate LSTM
        out, (h, c) = self.lstm(x, h)
        
        # Reshape output to (batch_size*sequence_length, hidden_size)
        out = out.reshape(out.size(0)*out.size(1), out.size(2))
        
        # Decode hidden states of all time steps
        out = self.linear(out)
        return out, (h, c)

model = RNN(vocab_size, embed_size, hidden_size, num_layers).to(device)
print(model)
````

<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Define Loss and Optimizer</h3>
</div>

````python
# Loss and Optimizer
criterion = nn.CrossEntropyLoss() 
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
````

````python
# Truncated backpropagation
def detach(states):
    return [state.detach() for state in states] 
````

<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h3 style='margin:10px 5px'>Model Training</h3>
</div>

````python
# Train the model
for epoch in range(num_epochs):
    # Set initial hidden and cell states
    states = (torch.zeros(num_layers, batch_size, hidden_size).to(device),
              torch.zeros(num_layers, batch_size, hidden_size).to(device))
    
    for i in range(0, ids.size(1) - seq_length, seq_length):
        # Get mini-batch inputs and targets
        inputs = ids[:, i:i+seq_length].to(device)
        targets = ids[:, (i+1):(i+1)+seq_length].to(device)
        
        # Forward pass
        states = detach(states)
        outputs, states = model(inputs, states)
        loss = criterion(outputs, targets.reshape(-1))
        
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        clip_grad_norm_(model.parameters(), 0.5)
        optimizer.step()

        step = (i+1) // seq_length
        if step % 100 == 0:
            print ('Epoch [{}/{}], Step[{}/{}], Loss: {:.4f}, Perplexity: {:5.2f}'
                   .format(epoch+1, num_epochs, step, num_batches, loss.item(), np.exp(loss.item())))
#Perplexity is an intrinsic measurement of how well a probability model predicts a sample
````

````python
# Test the model
with torch.no_grad():
    with open('sample.txt', 'w') as f:
        # Set intial hidden ane cell states
        state = (torch.zeros(num_layers, 1, hidden_size).to(device),
                 torch.zeros(num_layers, 1, hidden_size).to(device))

        # Select one word id randomly
        prob = torch.ones(vocab_size)
        input = torch.multinomial(prob, num_samples=1).unsqueeze(1).to(device)

        for i in range(num_samples):
            # Forward propagate RNN 
            output, state = model(input, state)

            # Sample a word id
            prob = output.exp()
            word_id = torch.multinomial(prob, num_samples=1).item()

            # Fill input with sampled word id for the next time step
            input.fill_(word_id)

            # File write
            word = corpus.dictionary.idx2word[word_id]
            word = '\n' if word == '<eos>' else word + ' '
            f.write(word)

            if (i+1) % 100 == 0:
                print('Sampled [{}/{}] words and save to {}'.format(i+1, num_samples, 'sample.txt'))
````
