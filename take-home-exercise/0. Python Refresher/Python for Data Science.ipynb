### Classes

**Class** is used to "create" objects.

```python
class my_class:
    #__init__ method is called every time a new object is created from the class
    def __init__(self, country):
        #class properties - all objects created from the class have access to this variable
        self.country = country
        print(f"Hello {self.country}!")
    
    #create a function that belongs to the class - this is called a method
    def talk(self):
        print(f"How's the JOSH {self.country}?")
```

```python
my_class_inst = my_class("India")
```

```python
my_class_inst.talk()
```

**Class inheritance** means that when you create a new class, you inherit the methods and properties of another class.

```python
#Example 1
class my_class_2(my_class):
    def talk(self):
        print(f"Can't hear it, a little louder {self.country}!")
```

```python
my_class_inst_2 = my_class_2("India")
```

```python
my_class_inst_2.talk()
```

```python
#Example 2
class my_class_3(my_class):
    def talk(self):
        super().talk()
        print(f"Can't hear it, a little louder {self.country}!")
```

```python
my_class_inst_3 = my_class_3("India")
```

```python
my_class_inst_3.talk()
```

## Other Useful Syntax

```python
#Scatterplot - Example 3
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(15,15))

#Set color map for different colors
colors = {'compact':'red', 'midsize':'green', 'suv':'blue', '2seater':'yellow', 'minivan': 'orange', \
          'pickup': 'purple', "subcompact": "pink"}

classes = ['compact', 'midsize', 'suv', '2seater', 'minivan', 'pickup', 'subcompact']

#Generate plot for highway miles vs. displacement for each type of car in different plot
for i in range(3):
    for j in range(3):
        ax[i][j].scatter(x="displ", y = "hwy", data=mpg_df.loc[(mpg_df["class"] == classes[3*i+j])], color = colors[classes[3*i+j]])    
        ax[i][j].set_title(classes[3*i+j])
        ax[i][j].set_xlabel("displ")
        ax[i][j].set_ylabel("hwy")
        
        #Define break condition
        if (3*i + j) == 6:
            break
    if (3*i + j) == 6:
        break
plt.tight_layout()
```

### Joint Distribution Plot

**Joint Distribution Plot** helps succintly see the bivariate relationship between 2 variables, and the individual distributions of the 2 variables.

### Heatmap

### Correlogram

**Correlogram** allow to understand correlation between all possible pairs of numeric data in the dataset.

```python
subset_cols = ["displ", "cyl", "cty", "hwy"]
```

```python
#Correlogram - Example 1
#Create figure canvas
plt.figure(figsize=(15,5))
sns.heatmap(mpg_df[subset_cols].corr(), xticklabels=mpg_df[subset_cols].corr().columns, yticklabels=mpg_df[subset_cols].corr().columns, center=0, annot=True)

#Aesthetics
plt.title('Correlogram of MPG', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
#Correlogram - Example 2
#Create figure canvas
plt.figure(figsize=(15,5))

#Create mask
mask = np.zeros_like(mpg_df[subset_cols].corr())
mask[np.triu_indices_from(mask)] = True

#Create Correlogram
sns.heatmap(mpg_df[subset_cols].corr(), xticklabels=mpg_df[subset_cols].corr().columns, \
            yticklabels=mpg_df[subset_cols].corr().columns, center=0, annot=True, mask=mask,\
            square=True, cmap="YlGnBu")

#Aesthetics
plt.title('Correlogram of MPG', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

### Strip Plots

**Stripplot** allows to see overlapping data points with same X and Y values, by slightly *jittering* the points.

### Box Plots

**Boxplots** are used to visualize the distribution, while keeping information about the outliers, 25th, 75th quartiles and outliers.

```python
###Lower bound = 25th percentile - 1.5*Interquartile range
###Upper bound = 75th percentile + 1.5*Interquartile range
# Interquartile range = Value at 75th percentile - Value at 25th percentile
##Anything beyond the bounds is an outlier
```

```python
#Boxplot - Example 1
plt.figure(figsize=(15,10))
sns.boxplot(x='class', y='hwy', data=mpg_df, notch=False)   

#Aesthetics
plt.title('Box Plot of Highway Mileage by Vehicle Class', fontsize=16)
plt.xlabel("Type of Car")
plt.ylabel("Highway Miles per Gallon")
plt.ylim(10, 40)
```

```python
#Boxplot - Example 1
plt.figure(figsize=(13,10), dpi= 80)
sns.boxplot(x='class', y='hwy', data=mpg_df, hue='cyl')
sns.stripplot(x='class', y='hwy', data=mpg_df, color='black', size=3, jitter=1)

#Aesthetics
plt.title('Boxplot of Highway Mileage by Vehicle Class', fontsize=16)
plt.xlabel("Type of Car")
plt.ylabel("Highway Miles per Gallon")
plt.legend(title='Cylinders')
```

### Violin Plots

**Violin plots** contain similar information to boxplots, but are visually appealing. The shape of violin depends on number of observations held.
