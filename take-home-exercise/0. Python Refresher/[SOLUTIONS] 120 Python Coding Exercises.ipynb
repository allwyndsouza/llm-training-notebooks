**Exercise 26** <br/>
Create a `Circle` class constructed using the diameter that can return the area and perimeter. Print area and perimeter of circle with diameter 10.

```python
#Solution 26
class my_circle():
    def __init__(self,diameter):
        self.diameter = diameter
    def circle_area(self):
        return(0.25*math.pi*(self.diameter**2))
    def circle_perimeter(self):
        return(math.pi*self.diameter)
```

```python
my_circle_instance = my_circle(10)
```

```python
print(my_circle_instance.circle_area())
```

```python
print(my_circle_instance.circle_perimeter())
```

**Exercise 27** <br/>
Create a `Cylinder` class inheriting from previously created `Circle` class, that is constructed using the diameter and height. It should n return the area and perimeter. Print area and perimeter of cylinder with diameter 10 and height 2.

```python
#Solution 27
class my_cyl(my_circle):
    def __init__(self,diameter,height):
        super().__init__(diameter)
        self.height = height
    def cyl_area(self):
        return(super().circle_area() * self.height)
    def cyl_perimeter(self):
        return(super().circle_perimeter() * self.height)
```

```python
my_cyl_instance = my_cyl(10,2)
```

```python
print(my_cyl_instance.cyl_area())
```

```python
print(my_cyl_instance.cyl_perimeter())
```

**Exercise 29** <br/>
Create a function to check if a triangle has exactly 3 sides equal, 2 sides equal or no sides equal to each other. Test it on triangle with sides `(4,5,5)`.

```python
#Solution 29
def my_func_7(sides_tup):
    side1 = sides_tup[0]
    side2 = sides_tup[1]
    side3 = sides_tup[2]
    if side1 == side2 == side3:
        return("All 3 sides equal")
    elif (side1 == side2) or (side2 == side3) or (side1 == side3):
        return("Exactly 2 sides equal")
    else:
        return("No sides are equal")
```

```python
my_tup = (4,5,5)
my_func_7(my_tup)
```
